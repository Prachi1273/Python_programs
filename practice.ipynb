```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10)
y=np.log10(x)
plt.plot(x,y)
plt.show()
```

```python
#f(x)=sin(x)-e^x+3X^2-log10(x) on interval [0,2pi]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,2*(np.pi),0.1)
y=np.sin(x)-np.exp(x)+3*x**2-np.log10(x)
plt.plot(x,y)
plt.show()
```
