```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import *
```

```python
import matplotlib.pyplot as plt
def f(x,y):
    return np.exp((-x)**2)
x=np.linspace(-5,5)
y=np.linspace(-5,5)
X,Y=np.meshgrid(x,y)
z=f(X,Y)
ax=axes(projection='3d')
ax.plot_surface(X,Y,Z)
legend()
show()
```

```python
#Write a Python program to generate 3D plot of the functions z = sin x + cos y in âˆ’10 < x, y < 10.
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y) :
    return np.sin(x)+np.cos(y)
x=np.linspace(-10,10)
y=np.linspace(-10,10)
X,Y=np.meshgrid(x,y)
z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,z,40)
legend()
show()
```

```python
#Using Python plot the surface plot of parabola z = x**2 + y**2 in âˆ’6 < x, y < 6
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from pylab import*
def f(x,y) :
    return x**2+y**2
x=np.linspace(-6,6,30)
y=np.linspace(-6,6,30)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
legend()
show()
```

```python
#Using Python plot the surface plot of function z = cos(x**2+y**2-0.5) in the interval from âˆ’1 < x, y < 1
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import*
def f(x,y):
    return np.cos(x**2+y**2-0.5)
x=np.linspace(-1,1)
y=np.linspace(-1,1)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.contour3D(X,Y,Z,50)
legend()
show()
```

```python
#) Generate 3D surface Plot for the function f(x) = sin(x**2+y**2) in the interval [0, 10]
```

```python
from mpl_toolkits import mplot3d
import numpy as np
from pylab import *
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))
x=np.linspace(0,10)
y=np.linspace(0,10)
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
ax=axes(projection='3d')
ax.plot_surface(X,Y,Z)
legend()
show()
```
